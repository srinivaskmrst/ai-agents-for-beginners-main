# бһҖбһ¶бһҡбһ„бһ¶бһҡбһҳбһ“бһ»бһҹбҹ’бһҹбһҖбҹ’бһ“бһ»бһ„бһӣбҹҶбһ“бһ¶бҹҶ (Human-in-the-Loop) бһҮбһ¶бһҳбһҪбһҷ Microsoft Agent Framework

## рҹҺҜ бһӮбҹ„бһӣбһҠбҹ…бһҖбһ¶бһҡбһҹбһ·бһҖбҹ’бһҹбһ¶

бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбҹҶбһҺбһҸбҹӢбһҸбҹ’бһҡбһ¶бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бһ–бһёбһҡбһ”бҹҖбһ”бһўбһ“бһ»бһңбһҸбҹ’бһҸбһ…бҹ’бһҡбһҖбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ **human-in-the-loop** бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ `RequestInfoExecutor` бһҡбһ”бһҹбҹӢ Microsoft Agent Frameworkбҹ” бһӮбҹҶбһҡбһјбһҠбҹӮбһӣбһҳбһ¶бһ“бһҗбһ¶бһҳбһ–бһӣбһ“бҹҒбҹҮбһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһұбҹ’бһҷбһўбҹ’бһ“бһҖбһ•бҹ’бһўбһ¶бһҖбҹӢбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ AI бһҠбһҫбһҳбҹ’бһ”бһёбһ”бҹ’бһҡбһҳбһјбһӣбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹ бһҠбҹӮбһӣбһ’бҹ’бһңбһҫбһұбҹ’бһҷбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһ”бһ¶бһ“ бһ“бһ·бһ„бһ•бҹ’бһҸбһӣбҹӢбһўбҹҶбһҺбһ¶бһ…бһҠбһӣбҹӢбһҳбһ“бһ»бһҹбҹ’бһҹбһӣбһҫбһҖбһ¶бһҡбһҹбһҳбҹ’бһҡбҹҒбһ…бһ…бһ·бһҸбҹ’бһҸбһҹбҹҶбһҒбһ¶бһ“бҹӢбҹ—бҹ”

## рҹ”„ бһҸбһҫ Human-in-the-Loop бһҮбһ¶бһўбҹ’бһңбһё?

**Human-in-the-loop (HITL)** бһӮбһәбһҮбһ¶бһӮбҹҶбһҡбһјбһҡбһ…бһ“бһ¶бһҠбҹӮбһӣбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡ AI бһ•бҹ’бһўбһ¶бһҖбҹӢбһҖбһ¶бһҡбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҠбһҫбһҳбҹ’бһ”бһёбһҹбҹ’бһ“бһҫбһҹбһ»бҹҶбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹбһҳбһ»бһ“бһ–бҹҒбһӣбһ”бһ“бҹ’бһҸбҹ” бһңбһ¶бһҹбҹҶбһҒбһ¶бһ“бҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбҹ–

- вң… **бһҹбҹҒбһ…бһҖбҹ’бһҸбһёбһҹбһҳбҹ’бһҡбҹҒбһ…бһҹбҹҶбһҒбһ¶бһ“бҹӢ** - бһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бһҖбһ¶бһҡбһўбһ“бһ»бһҳбҹҗбһҸбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹбһҳбһ»бһ“бһ’бҹ’бһңбһҫбһҹбһҖбһҳбҹ’бһҳбһ—бһ¶бһ–бһҹбҹҶбһҒбһ¶бһ“бҹӢбҹ—
- вң… **бһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бһҳбһ·бһ“бһ…бҹ’бһ”бһ¶бһҹбҹӢбһӣбһ¶бһҹбҹӢ** - бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһұбҹ’бһҷбһҳбһ“бһ»бһҹбҹ’бһҹбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһ–бҹҒбһӣ AI бһҳбһ·бһ“бһ”бҹ’бһҡбһ¶бһҖбһҠ
- вң… **бһ…бҹҶбһҺбһ„бҹӢбһ…бҹҶбһҺбһјбһӣбһ…бһ·бһҸбҹ’бһҸбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ** - бһҹбһҪбһҡбһІбҹ’бһҷбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҡбһңбһ¶бһ„бһҮбһҳбҹ’бһҡбһҫбһҹбһ…бҹ’бһҡбһҫбһ“
- вң… **бһҖбһ¶бһҡбһўбһ“бһ»бһңбһҸбҹ’бһҸбһҸбһ¶бһҳбһ”бһ‘бһ”бһүбҹ’бһҮбһ¶ бһ“бһ·бһ„бһҹбһ»бһңбһҸбҹ’бһҗбһ·бһ—бһ¶бһ–** - бһҠбһҫбһҳбҹ’бһ”бһёбһ’бһ¶бһ“бһ¶бһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҠбҹӮбһӣбһҳбһ¶бһ“бһ”бһ‘бһ”бһүбҹ’бһҮбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„
- вң… **бһ”бһ‘бһ–бһ·бһҹбҹ„бһ’бһ“бҹҚбһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳ** - бһ”бһ„бҹ’бһҖбһҫбһҸбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһ–бһ·бһ—бһ¶бһҖбҹ’бһҹбһ¶бһҠбҹӮбһӣбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһ‘бҹ…бһ“бһ№бһ„бһ”бһүбҹ’бһ…бһјбһӣбһ–бһёбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ

## рҹҸ—пёҸ бһңбһ¶бһҸбҹ’бһҡбһјбһңбһ”бһҫбһҖбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„ Microsoft Agent Framework

Framework бһ“бҹҒбҹҮбһ•бҹ’бһҸбһӣбҹӢбһ§бһ”бһҖбһҡбһҺбҹҚбһҹбҹҶбһҒбһ¶бһ“бҹӢбһ”бһёбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ HITL:

1. **`RequestInfoExecutor`** - бһҮбһ¶бһўбҹ’бһ“бһҖбһўбһ“бһ»бһңбһҸбҹ’бһҸбһ–бһ·бһҹбҹҒбһҹбһҠбҹӮбһӣбһ”бһүбҹ’бһҲбһ”бҹӢбһ…бһҡбһ“бҹ’бһҸбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ бһ бһҫбһҷбһ”бһүбҹ’бһ…бҹҒбһү `RequestInfoEvent`
2. **`RequestInfoMessage`** - бһҗбҹ’бһ“бһ¶бһҖбҹӢбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ‘бһҳбҹ’бһҡбһ„бҹӢбһҹбҹҶбһҺбһҫбһҠбҹӮбһӣбһҳбһ¶бһ“бһ”бҹ’бһҡбһ—бҹҒбһ‘ бһҠбҹӮбһӣбһ•бҹ’бһүбһҫбһ‘бҹ…бһҳбһ“бһ»бһҹбҹ’бһҹ
3. **`RequestResponse`** - бһ—бҹ’бһҮбһ¶бһ”бҹӢбһ…бһҳбҹ’бһӣбһҫбһҷбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹбһҮбһ¶бһҳбһҪбһҷбһҹбҹҶбһҺбһҫбһҠбһҫбһҳбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ `request_id`

**Workflow Pattern:**
```
Agent detects need for input
    вҶ“
Sends message to RequestInfoExecutor
    вҶ“
Workflow pauses & emits RequestInfoEvent
    вҶ“
Application collects human input (console, UI, etc.)
    вҶ“
Application sends RequestResponse via send_responses_streaming()
    вҶ“
Workflow resumes with human input
```

## рҹҸЁ бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһҡбһ”бһҹбҹӢбһҷбһҫбһ„: бһҖбһ¶бһҡбһҖбһҖбҹӢбһҹбһҺбҹ’бһӢбһ¶бһӮбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷбһҖбһ¶бһҡбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһ–бһёбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ

бһҷбһҫбһ„бһ“бһ№бһ„бһҹбҹ’бһҗбһ¶бһ”бһ“бһ¶бһЎбһҫбһ„бһӣбһҫбһ…бҹ’бһҡбһҖбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡвҖӢбһҖбҹ’бһ“бһ»бһ„бһӣбһҖбҹ’бһҒбһҒбһҺбҹ’бһҢ бһҠбҹ„бһҷбһ”бһ“бҹ’бһҗбҹӮбһҳбһҖбһ¶бһҡбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹ **бһҳбһ»бһ“** бһҖбһ¶бһҡбһ•бҹ’бһҸбһӣбҹӢбһўбһ“бһ»бһҹбһ¶бһҹбһ“бҹҚбһ–бһёбһ‘бһёбһҖбһ“бҹ’бһӣбҹӮбһ„бһҮбҹҶбһ“бһҪбһҹбҹ–

1. бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһҹбҹ’бһ“бһҫбһҡбһҹбһ»бҹҶбһ‘бһёбһҖбһ“бҹ’бһӣбҹӮбһ„ (бһ§. "Paris")
2. `availability_agent` бһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбһҫбһӣбһҗбһ¶бһҸбһҫбһ”бһ“бҹ’бһ‘бһ”бҹӢбһҳбһ¶бһ“бһ“бҹ…бһ‘бҹҒ
3. **If no rooms** вҶ’ `confirmation_agent` бһҹбһҪбһҡ "бһҸбһҫбһўбҹ’бһ“бһҖбһ…бһ„бҹӢбһҳбһҫбһӣбһҮбһҳбҹ’бһҡбһҫбһҹбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһ¬?"
4. Workflow **pauses** using `RequestInfoExecutor`
5. **Human responds** "yes" or "no" via console input
6. `decision_manager` бһ”бһүбҹ’бһҮбһјбһ“бһҸбһ¶бһҳбһ…бһҳбҹ’бһӣбһҫбһҷ:
   - **Yes** вҶ’ бһ”бһ„бҹ’бһ бһ¶бһүбһ‘бһёбһҖбһ“бҹ’бһӣбҹӮбһ„бһҮбҹҶбһ“бһҪбһҹ
   - **No** вҶ’ бһ”бҹ„бҹҮбһ”бһ„бҹӢбһҹбҹҶбһҺбһҫбһҖбһҖбҹӢ
7. бһ”бһ„бҹ’бһ бһ¶бһүбһӣбһ‘бҹ’бһ’бһ•бһӣбһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷ

бһ“бҹҒбҹҮбһ”бһ„бҹ’бһ бһ¶бһүбһ–бһёбһҡбһ”бҹҖбһ”бһ•бҹ’бһҸбһӣбҹӢбһўбҹҶбһҺбһ¶бһ…бһҠбһӣбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһӣбһҫбһўбһ“бһ»бһҹбҹ’бһҹбһ¶бһҡбһҺбҹҲбһҡбһ”бһҹбҹӢбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡ!

## бһҮбҹҶбһ бһ¶бһ“ 1: бһ“бһ¶бҹҶбһ…бһјбһӣбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡ

бһҷбһҫбһ„бһ“бһ¶бҹҶбһ…бһјбһӣбһ’бһ¶бһҸбһ»бһ•бҹ’бһҹбҹҶбһҹбҹ’бһҸбһ„бҹӢбһҠбһ¶бһҡбһҡбһ”бһҹбҹӢ Agent Framework бһ”бһјбһҖбһҡбһҪбһҳбһ“бһ№бһ„ **бһҗбҹ’бһ“бһ¶бһҖбҹӢбһ–бһ·бһҹбҹҒбһҹбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҳбһ“бһ»бһҹбҹ’бһҹбһҖбҹ’бһ“бһ»бһ„бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ**:
- `RequestInfoExecutor` - Executor бһҠбҹӮбһӣбһ•бҹ’бһўбһ¶бһҖбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹ
- `RequestInfoEvent` - бһ–бҹ’бһҡбһ№бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҺбҹҚбһҠбҹӮбһӣбһ”бһүбҹ’бһ…бҹҒбһүбһ–бҹҒбһӣбһҠбҹӮбһӣбһҳбһ¶бһ“бһҖбһ¶бһҡбһҹбҹ’бһ“бһҫбһҹбһ»бҹҶбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹ
- `RequestInfoMessage` - бһҗбҹ’бһ“бһ¶бһҖбҹӢбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҹбҹҶбһҺбһҫбһҠбҹӮбһӣбһҳбһ¶бһ“бһ”бҹ’бһҡбһ—бҹҒбһ‘
- `RequestResponse` - бһ—бҹ’бһҮбһ¶бһ”бҹӢбһҖбһ¶бһҡбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹбһҮбһ¶бһҳбһҪбһҷбһҹбҹҶбһҺбһҫ
- `WorkflowOutputEvent` - бһ–бҹ’бһҡбһ№бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҺбҹҚбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҡбһҖбһғбһҫбһүбһӣбһ‘бҹ’бһ’бһ•бһӣбһ–бһёбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ

## бһҮбҹҶбһ бһ¶бһ“бһ‘бһё 2: бһҖбҹҶбһҺбһҸбҹӢбһҳбҹүбһјбһҢбҹӮбһӣ Pydantic бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһӣбһ‘бҹ’бһ’бһ•бһӣбһҠбҹӮбһӣбһҳбһ¶бһ“бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’

бһҳбҹүбһјбһҢбҹӮбһӣбһ‘бһ¶бҹҶбһ„бһ“бҹҒбҹҮбһҖбҹҶбһҺбһҸбҹӢ **бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’** бһҠбҹӮбһӣбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһ“бһ№бһ„бһҸбҹ’бһҡбһЎбһ”бҹӢбһҳбһҖбҹ” бһҷбһҫбһ„бһҡбһҖбҹ’бһҹбһ¶бһҳбҹүбһјбһҢбҹӮбһӣбһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢбһ–бһёбһ…бһҡбһ“бҹ’бһҸбһҖбһ¶бһҡбһ„бһ¶бһҡбһҸбһ¶бһҳбһӣбһҖбҹ’бһҒбһҒбһҺбҹ’бһҢ бһ бһҫбһҷбһ”бһ“бҹ’бһҗбҹӮбһҳ:

**бһҗбҹ’бһҳбһёбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҳбһ“бһ»бһҹбҹ’бһҹбһҖбҹ’бһ“бһ»бһ„бһҹбҹҖбһӮбҹ’бһңбһё (Human-in-the-Loop):**
- `HumanFeedbackRequest` - бһҮбһ¶бһҖбһјбһ“бһҗбҹ’бһ“бһ¶бһҖбҹӢбһ“бҹғ `RequestInfoMessage` бһҠбҹӮбһӣбһҖбҹҶбһҺбһҸбҹӢбһ‘бһҳбҹ’бһҡбһ„бҹӢбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҹбҹҶбһҺбһҫбһҠбҹӮбһӣбһ•бҹ’бһүбһҫбһ‘бҹ…бһҳбһ“бһ»бһҹбҹ’бһҹ
  - бһҳбһ¶бһ“ `prompt` (бһҹбҹҶбһҺбһҪбһҡбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҹбһҪбһҡ) бһ“бһ·бһ„ `destination` (бһ”бһҡбһ·бһ”бһ‘бһўбҹҶбһ–бһёбһ‘бһёбһҖбҹ’бһҡбһ»бһ„бһҠбҹӮбһӣбһҳбһ·бһ“бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ”бһ¶бһ“)


````python
# Existing models from conditional workflow
class BookingCheckResult(BaseModel):
    """Result from checking hotel availability at a destination."""
    destination: str
    has_availability: bool
    message: str


class AlternativeResult(BaseModel):
    """Suggested alternative destination when no rooms available."""
    alternative_destination: str
    reason: str


class BookingConfirmation(BaseModel):
    """Booking suggestion when rooms are available."""
    destination: str
    action: str
    message: str


# NEW: Pydantic model for agent's response format
class ConfirmationQuestion(BaseModel):
    """
    Pydantic model used by confirmation_agent's response_format.
    This is what the agent will output as JSON.
    """
    question: str  # The question to ask the user
    destination: str  # The unavailable destination for context


# NEW: Dataclass for RequestInfoExecutor
@dataclass
class HumanFeedbackRequest(RequestInfoMessage):
    """
    Request sent to RequestInfoExecutor asking if user wants alternatives.
    
    MUST be a dataclass subclassing RequestInfoMessage for type compatibility.
    This is what gets sent to the RequestInfoExecutor.
    """
    prompt: str = ""  # The question to ask the user
    destination: str = ""  # The unavailable destination for context


print("вң… Pydantic models defined:")
print("   - BookingCheckResult (availability check)")
print("   - AlternativeResult (alternative suggestion)")
print("   - BookingConfirmation (booking confirmation)")
print("   - ConfirmationQuestion (agent response format) рҹҶ•")
print("   - HumanFeedbackRequest (RequestInfoMessage for HITL) рҹҶ•")
````

## бһҮбҹҶбһ бһ¶бһ“бһ‘бһё 3: бһ”бһ„бҹ’бһҖбһҫбһҸбһ§бһ”бһҖбһҡбһҺбҹҚбһҖбһҖбҹӢбһҹбһҺбҹ’бһӢбһ¶бһӮбһ¶бһҡ

бһ§бһ”бһҖбһҡбһҺбҹҚбһҠбһјбһ…бһӮбҹ’бһ“бһ¶бһ–бһёбһӣбҹҶбһ бһјбһҡбһҖбһ¶бһҡбһ„бһ¶бһҡбһҸбһ¶бһҳбһӣбһҖбҹ’бһҒбһҒбһҺбҹ’бһҢ - бһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбһҫбһӣбһҗбһ¶бһҸбһҫбһҳбһ¶бһ“бһ”бһ“бҹ’бһ‘бһ”бҹӢбһ‘бҹҶбһ“бҹҒбһҡбһ“бҹ…бһӮбҹ„бһӣбһҠбҹ…бһ¬бһ‘бҹҒбҹ”


````python
@ai_function(description="Check hotel room availability for a destination city")
def hotel_booking(destination: Annotated[str, "The destination city to check for hotel rooms"]) -> str:
    """
    Simulates checking hotel room availability.
    
    Returns JSON string with availability status.
    """
    display(
        HTML(f"""
        <div style='padding: 15px; background: #e3f2fd; border-left: 4px solid #2196f3; border-radius: 4px; margin: 10px 0;'>
            <strong>рҹ”Қ Tool Invoked:</strong> hotel_booking("{destination}")
        </div>
    """)
    )

    # Simulate availability check
    cities_with_rooms = ["stockholm", "seattle", "tokyo", "london", "amsterdam"]
    has_rooms = destination.lower() in cities_with_rooms

    result = {"has_availability": has_rooms, "destination": destination}

    return json.dumps(result)


print("вң… hotel_booking tool created with @ai_function decorator")
````

## бһҮбҹҶбһ бһ¶бһ“ 4: бһҖбҹҶбһҺбһҸбҹӢбһҳбһ»бһҒбһ„бһ¶бһҡвҖӢбһӣбһҖбҹ’бһҒбһҒбһҺбҹ’бһҢвҖӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҸбһҳбҹ’бһҡбһ„

бһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡ **бһҳбһ»бһҒбһ„бһ¶бһҡвҖӢбһӣбһҖбҹ’бһҒбһҒбһҺбҹ’бһҢвҖӢбһ…бҹҶбһ“бһҪбһ“вҖӢбһ”бһҪбһ“** бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡвҖӢбһҠбҹӮбһӣбһҳбһ¶бһ“бһҳбһ“бһ»бһҹбҹ’бһҹбһ…бһјбһӣбһҡбһҪбһҳбһҡбһ”бһҹбҹӢбһҷбһҫбһ„:

**бһ–бһёбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡвҖӢбһҸбһ¶бһҳбһӣбһҖбҹ’бһҒбһҒбһҺбҹ’бһҢ:**
1. `has_availability_condition` - бһ”бһүбҹ’бһҮбһјбһ“бһ–бҹҒбһӣбһҠбҹӮбһӣбһҹбһҺбҹ’бһӢбһ¶бһӮбһ¶бһҡбһҳбһ¶бһ“бһ‘бҹҶбһ“бҹҒбһҡ
2. `no_availability_condition` - бһ”бһүбҹ’бһҮбһјбһ“бһ–бҹҒбһӣбһҠбҹӮбһӣбһҹбһҺбҹ’бһӢбһ¶бһӮбһ¶бһҡбһҳбһ·бһ“бһҳбһ¶бһ“бһ‘бҹҶбһ“бҹҒбһҡ

**бһҗбҹ’бһҳбһёбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡвҖӢбһҠбҹӮбһӣбһҳбһ¶бһ“бһҳбһ“бһ»бһҹбҹ’бһҹбһ…бһјбһӣбһҡбһҪбһҳ:**
3. `user_wants_alternatives_condition` - бһ”бһүбҹ’бһҮбһјбһ“бһ–бҹҒбһӣбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҷбһӣбҹӢвҖӢбһ–бҹ’бһҡбһҳбһ…бҹҶбһ–бҹ„бҹҮбһҮбһҳбҹ’бһҡбһҫбһҹбһҮбҹҶбһ“бһҪбһҹ
4. `user_declines_alternatives_condition` - бһ”бһүбҹ’бһҮбһјбһ“бһ–бҹҒбһӣбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһ”бһҠбһ·бһҹбҹҒбһ’бһҮбһҳбҹ’бһҡбһҫбһҹбһҮбҹҶбһ“бһҪбһҹ


````python
# Existing condition functions from conditional workflow
def has_availability_condition(message: Any) -> bool:
    """Condition for routing when hotels ARE available."""
    if not isinstance(message, AgentExecutorResponse):
        return True

    try:
        result = BookingCheckResult.model_validate_json(message.agent_run_response.text)
        display(
            HTML(f"""
            <div style='padding: 12px; background: #c8e6c9; border-left: 4px solid #4caf50; border-radius: 4px; margin: 10px 0;'>
                <strong>вң… Condition Check:</strong> has_availability = <strong>{result.has_availability}</strong> for {result.destination}
            </div>
        """)
        )
        return result.has_availability
    except Exception as e:
        display(HTML(f"""<div style='padding: 12px; background: #ffcdd2; border-left: 4px solid #f44336; border-radius: 4px; margin: 10px 0;'><strong>вҡ пёҸ  Error:</strong> {str(e)}</div>"""))
        return False


def no_availability_condition(message: Any) -> bool:
    """Condition for routing when hotels are NOT available."""
    if not isinstance(message, AgentExecutorResponse):
        return False

    try:
        result = BookingCheckResult.model_validate_json(message.agent_run_response.text)
        display(
            HTML(f"""
            <div style='padding: 12px; background: #ffecb3; border-left: 4px solid #ff9800; border-radius: 4px; margin: 10px 0;'>
                <strong>вқҢ Condition Check:</strong> no_availability for {result.destination}
            </div>
        """)
        )
        return not result.has_availability
    except Exception as e:
        return False


# NEW: Condition functions for human-in-the-loop routing
def user_wants_alternatives_condition(message: Any) -> bool:
    """
    Condition for routing when user WANTS to see alternatives.
    
    Checks the AgentExecutorRequest sent by decision_manager.
    """
    # Check if it's an AgentExecutorRequest (what decision_manager sends)
    if isinstance(message, AgentExecutorRequest):
        # Check the message text to determine user's choice
        if message.messages and len(message.messages) > 0:
            msg_text = message.messages[0].text.lower()
            wants_alternatives = "wants to see alternative" in msg_text or "want to see alternative" in msg_text
            
            display(
                HTML(f"""
                <div style='padding: 12px; background: #e1f5fe; border-left: 4px solid #0288d1; border-radius: 4px; margin: 10px 0;'>
                    <strong>рҹ”Қ User Decision:</strong> User wants alternatives = <strong>{wants_alternatives}</strong>
                </div>
            """)
            )
            
            return wants_alternatives
    
    return False
def user_declines_alternatives_condition(message: Any) -> bool:
    """
    Condition for routing when user DECLINES alternatives.
    
    Checks the AgentExecutorRequest sent by decision_manager.
    """
    # Check if it's an AgentExecutorRequest (what decision_manager sends)
    if isinstance(message, AgentExecutorRequest):
        # Check the message text to determine user's choice
        if message.messages and len(message.messages) > 0:
            msg_text = message.messages[0].text.lower()
            declined = "declined" in msg_text or "has declined" in msg_text
            
            display(
                HTML(f"""
                <div style='padding: 12px; background: #fce4ec; border-left: 4px solid #c2185b; border-radius: 4px; margin: 10px 0;'>
                    <strong>рҹҡ« User Decision:</strong> User declined alternatives = <strong>{declined}</strong>
                </div>
            """)
            )
            
            return declined
    
    return False
print("вң… Condition functions defined:")
print("   - has_availability_condition (routes when rooms exist)")
print("   - no_availability_condition (routes when no rooms)")
print("   - user_wants_alternatives_condition (routes when user says yes) рҹҶ•")
print("   - user_declines_alternatives_condition (routes when user says no) рҹҶ•")
````

## бһҮбҹҶбһ бһ¶бһ“бһ‘бһё 5: бһ”бһ„бҹ’бһҖбһҫбһҸ Decision Manager Executor

бһ“бҹҒбҹҮбһӮбһәбһҮбһ¶ **бһӮбҹ„бһӣбһҹбҹ’бһ“бһјбһӣбһ“бҹғбһӮбҹҶбһҡбһј human-in-the-loop**! `DecisionManager` бһӮбһәбһҮбһ¶ `Executor` бһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“ бһҠбҹӮбһӣбҹ–

1. **бһ‘бһ‘бһҪбһӣбһҳбһҸбһ·бһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹ** бһҸбһ¶бһҳбһҡбһҷбҹҲ бһңбһҸбҹ’бһҗбһ» `RequestResponse`
2. **бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҹбҹҒбһ…бһҖбҹ’бһҸбһёбһҹбһҳбҹ’бһҡбҹҒбһ…бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ** (yes/no)
3. **бһҺбҹӮбһ“бһ¶бҹҶбһ‘бһ·бһҹбһҠбҹ…бһ“бҹғбһӣбҹҶбһ бһјбһҡбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ** бһҠбҹ„бһҷбһ•бҹ’бһүбһҫбһҹбһ¶бһҡбһ‘бҹ…бһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһҠбҹӮбһӣбһҹбһҳбһҡбһҳбҹ’бһҷ

Key features:
- бһ”бҹ’бһҡбһҫ `@handler` decorator бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһ бһ¶бһүбһҳбһ»бһҒбһ„бһ¶бһҡ (methods) бһҮбһ¶бһҮбҹҶбһ бһ¶бһ“бһ“бҹғбһӣбҹҶбһ бһјбһҡбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ
- бһ‘бһ‘бһҪбһӣ `RequestResponse[HumanFeedbackRequest, str]` бһҠбҹӮбһӣбһҳбһ¶бһ“бһ‘бһ¶бҹҶбһ„бһҹбҹҶбһҺбһҫбһҠбһҫбһҳ бһ“бһ·бһ„бһ…бһҳбҹ’бһӣбһҫбһҷбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ
- бһ”бһүбҹ’бһ…бҹҒбһүбһҹбһ¶бһҡвҖӢбһ„бһ¶бһҷбҹ— "yes" бһ¬ "no" бһҠбҹӮбһӣбһ”бһҺбҹ’бһҸбһ¶бһӣбһІбҹ’бһҷбһҳбһ»бһҒбһ„бһ¶бһҡбһӣбһҖбҹ’бһҒбһҒбһҺбҹ’бһҢбһҡбһ”бһҹбҹӢбһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ


````python
class DecisionManager(Executor):
    """
    Coordinates workflow routing based on human feedback.
    
    This executor receives RequestResponse objects from the RequestInfoExecutor
    and makes routing decisions by sending simple messages that trigger
    condition functions.
    """

    def __init__(self, id: str | None = None):
        super().__init__(id=id or "decision_manager")

    @handler
    async def on_human_feedback(
        self,
        feedback: RequestResponse[HumanFeedbackRequest, str],
        ctx: WorkflowContext[AgentExecutorRequest],
    ) -> None:
        """
        Process human feedback and let the workflow route based on conditions.
        
        The RequestResponse contains:
        - feedback.data: The user's string reply (e.g., "yes" or "no")
        - feedback.original_request: The HumanFeedbackRequest with context
        
        This handler just displays feedback and passes the RequestResponse through.
        The routing is done by condition functions on the edges.
        """
        user_reply = (feedback.data or "").strip().lower()
        destination = getattr(feedback.original_request, "destination", "unknown")

        display(
            HTML(f"""
            <div style='padding: 15px; background: #f3e5f5; border-left: 4px solid #9c27b0; border-radius: 4px; margin: 10px 0;'>
                <strong>рҹҺҜ Decision Manager:</strong> Processing user reply: <strong>"{user_reply}"</strong> for {destination}
            </div>
        """)
        )

        if user_reply == "yes":
            display(
                HTML("""
                <div style='padding: 12px; background: #c8e6c9; border-left: 4px solid #4caf50; border-radius: 4px; margin: 10px 0;'>
                    <strong>вһЎпёҸ  Routing:</strong> User wants alternatives вҶ’ Will route to alternative_agent
                </div>
            """)
            )
            # Create and send a message for the alternative_agent
            user_msg = ChatMessage(
                Role.USER,
                text=f"The user wants to see alternative destinations near {destination}. Please suggest one.",
            )
            await ctx.send_message(AgentExecutorRequest(messages=[user_msg], should_respond=True))
        
        elif user_reply == "no":
            display(
                HTML("""
                <div style='padding: 12px; background: #ffcdd2; border-left: 4px solid #f44336; border-radius: 4px; margin: 10px 0;'>
                    <strong>рҹҡ« Routing:</strong> User declined alternatives вҶ’ Will route to cancellation_agent
                </div>
            """)
            )
            # Create and send a message for the cancellation_agent
            user_msg = ChatMessage(
                Role.USER,
                text="The user has declined to see alternatives. Please acknowledge their decision.",
            )
            await ctx.send_message(AgentExecutorRequest(messages=[user_msg], should_respond=True))
        
        else:
            # Handle unexpected input - treat as decline
            display(
                HTML(f"""
                <div style='padding: 12px; background: #fff3e0; border-left: 4px solid #ff9800; border-radius: 4px; margin: 10px 0;'>
                    <strong>вҡ пёҸ  Warning:</strong> Unexpected input "{user_reply}" - treating as decline
                </div>
            """)
            )
            user_msg = ChatMessage(
                Role.USER,
                text="The user has declined to see alternatives. Please acknowledge their decision.",
            )
            await ctx.send_message(AgentExecutorRequest(messages=[user_msg], should_respond=True))


print("вң… DecisionManager executor created with @handler method for human feedback")
````

## бһҮбҹҶбһ бһ¶бһ“бһ‘бһё 6: бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһўбһ“бһ»бһңбһҸбҹ’бһҸбһ”бһ„бҹ’бһ бһ¶бһүбһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“

бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһўбһ“бһ»бһңбһҸбҹ’бһҸбһ”бһ„бҹ’бһ бһ¶бһүбһҠбһјбһ…бһӮбҹ’бһ“бһ¶бһ–бһё workflow бһҠбҹӮбһӣбһҳбһ¶бһ“бһӣбһҖбҹ’бһҒбһҒбһҺбҹ’бһҢ - бһ•бҹ’бһҠбһӣбҹӢбһӣбһ‘бҹ’бһ’бһ•бһӣбһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷбһҮбһ¶ output бһ“бҹғ workflowбҹ”


````python
@executor(id="prepare_human_request")
async def prepare_human_request(
    response: AgentExecutorResponse, 
    ctx: WorkflowContext[HumanFeedbackRequest]
) -> None:
    """
    Transform agent response into HumanFeedbackRequest for RequestInfoExecutor.
    
    This executor bridges the type gap between:
    - confirmation_agent outputs AgentExecutorResponse with ConfirmationQuestion JSON
    - request_info_executor expects HumanFeedbackRequest (RequestInfoMessage dataclass)
    """
    display(
        HTML("""
        <div style='padding: 12px; background: #e1f5fe; border-left: 4px solid #0288d1; border-radius: 4px; margin: 10px 0;'>
            <strong>рҹ”„ Transform:</strong> Converting ConfirmationQuestion to HumanFeedbackRequest
        </div>
    """)
    )
    
    # Parse the agent's Pydantic output (ConfirmationQuestion)
    confirmation = ConfirmationQuestion.model_validate_json(response.agent_run_response.text)
    
    # Convert to HumanFeedbackRequest dataclass for RequestInfoExecutor
    feedback_request = HumanFeedbackRequest(
        prompt=confirmation.question,
        destination=confirmation.destination
    )
    
    # Send the properly typed RequestInfoMessage to the RequestInfoExecutor
    await ctx.send_message(feedback_request)


@executor(id="display_result")
async def display_result(response: AgentExecutorResponse, ctx: WorkflowContext[Never, str]) -> None:
    """
    Display the final result as workflow output.
    
    This executor receives the final agent response and yields it as the workflow output.
    """
    display(
        HTML("""
        <div style='padding: 15px; background: #f3e5f5; border-left: 4px solid #9c27b0; border-radius: 4px; margin: 10px 0;'>
            <strong>рҹ“Ө Display Executor:</strong> Yielding workflow output
        </div>
    """)
    )

    await ctx.yield_output(response.agent_run_response.text)


print("вң… prepare_human_request executor created with @executor decorator")
print("вң… display_result executor created with @executor decorator")
````

бһҖбҹҶбһҺбһҸбҹӢбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһўбһҸбһ·бһҗбһ·бһҮбһ“ LLM (GitHub Models, Azure OpenAI, бһ¬ OpenAI).

## бһҮбҹҶбһ бһ¶бһ“ 8: бһ”бһ„бҹ’бһҖбһҫбһҸбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡ AI бһ“бһ·бһ„ бһўбҹ’бһ“бһҖбһўбһ“бһ»бһңбһҸбҹ’бһҸ

We create **бһҹбһҳбһ¶бһҹбһ’бһ¶бһҸбһ»бһ”бҹ’бһҡбһ¶бҹҶбһҳбһҪбһҷбһҡбһ”бһҹбҹӢбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ„бһ¶бһҡ**:

**бһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡ (бһҠбҹӮбһӣбһ”бһүбҹ’бһ…бһјбһӣбһҖбҹ’бһ“бһ»бһ„ AgentExecutor):**
1. **availability_agent** - бһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһ—бһ¶бһ–бһҳбһ¶бһ“бһ”бһ“бҹ’бһ‘бһ”бҹӢбһ“бҹ…бһҹбһҺбҹ’бһӢбһ¶бһӮбһ¶бһҡ бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһ§бһ”бһҖбһҡбһҺбҹҚ
2. **confirmation_agent** - рҹҶ• бһҡбҹҖбһ”бһ…бҹҶбһҹбҹҶбһҺбһҫбһҹбһ»бҹҶбһҖбһ¶бһҡбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹ
3. **alternative_agent** - бһҹбҹ’бһ“бһҫбһ‘бһёбһҖбҹ’бһҡбһ»бһ„бһҮбҹҶбһ“бһҪбһҹ (бһ“бҹ…бһ–бҹҒбһӣбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҶбҹ’бһӣбһҫбһҷбһҗбһ¶ 'бһҷбһӣбҹӢбһ–бҹ’бһҡбһҳ')
4. **booking_agent** - бһӣбһҫбһҖбһ‘бһ№бһҖбһ…бһ·бһҸбҹ’бһҸбһІбҹ’бһҷбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҖбһҖбҹӢ (бһ“бҹ…бһ–бҹҒбһӣбһҳбһ¶бһ“бһ”бһ“бҹ’бһ‘бһ”бҹӢ)
5. **cancellation_agent** - рҹҶ• бһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһҹбһ¶бһҡбһӣбһ»бһ”бһ…бҹ„бһӣ (бһ“бҹ…бһ–бҹҒбһӣбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҶбҹ’бһӣбһҫбһҷбһҗбһ¶ 'бһ‘бҹҒ')

**бһўбҹ’бһ“бһҖбһўбһ“бһ»бһңбһҸбҹ’бһҸбһ–бһ·бһҹбҹҒбһҹ:**
6. **request_info_executor** - рҹҶ• `RequestInfoExecutor` бһҠбҹӮбһӣбһ•бҹ’бһўбһ¶бһҖбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ„бһ¶бһҡбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹ
7. **decision_manager** - рҹҶ• бһўбҹ’бһ“бһҖбһўбһ“бһ»бһңбһҸбҹ’бһҸбһҸбһ¶бһҳбһҸбһҳбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҠбҹӮбһӣбһҡбҹҖбһ”бһ…бҹҶбһ‘бһ·бһҹбһҠбҹ…бһўбһ¶бһҹбҹ’бһҡбҹҗбһҷбһӣбһҫбһҖбһ¶бһҡбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹ (бһ”бһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢбһҡбһҪбһ…бһ“бҹ…бһҒбһ¶бһ„бһӣбһҫ)


````python
# Agent 1: Check availability with tool (same as conditional workflow)
availability_agent = AgentExecutor(
    chat_client.create_agent(
        instructions=(
            "You are a hotel booking assistant that checks room availability. "
            "Use the hotel_booking tool to check if rooms are available at the destination. "
            "Return JSON with fields: destination (string), has_availability (bool), and message (string). "
            "The message should summarize the availability status."
        ),
        tools=[hotel_booking],
        response_format=BookingCheckResult,
    ),
    id="availability_agent",
)

# Agent 2: NEW - Prepare human confirmation request
confirmation_agent = AgentExecutor(
    chat_client.create_agent(
        instructions=(
            "You are a helpful assistant. The user's requested destination has no available hotel rooms. "
            "Create a polite message asking if they would like to see alternative destinations nearby. "
            "Return a JSON with: destination (the unavailable city), and question (a friendly yes/no question). "
            "Keep the question concise and friendly."
        ),
        response_format=ConfirmationQuestion,  # Use Pydantic model for agent output
    ),
    id="confirmation_agent",
)

# Agent 3: Suggest alternative (when user says yes)
alternative_agent = AgentExecutor(
    chat_client.create_agent(
        instructions=(
            "You are a helpful travel assistant. When a user cannot find hotels in their requested city, "
            "suggest an alternative nearby city that has availability. "
            "Return JSON with fields: alternative_destination (string) and reason (string). "
            "Make your suggestion sound appealing and helpful."
        ),
        response_format=AlternativeResult,
    ),
    id="alternative_agent",
)

# Agent 4: Suggest booking (when rooms available)
booking_agent = AgentExecutor(
    chat_client.create_agent(
        instructions=(
            "You are a booking assistant. The user has found available hotel rooms. "
            "Encourage them to book by highlighting the destination's appeal. "
            "Return JSON with fields: destination (string), action (string), and message (string). "
            "The action should be 'book_now' and message should be encouraging."
        ),
        response_format=BookingConfirmation,
    ),
    id="booking_agent",
)

# Agent 5: NEW - Handle cancellation when user declines alternatives
class CancellationMessage(BaseModel):
    """Message when user declines alternatives."""
    status: str
    message: str

cancellation_agent = AgentExecutor(
    chat_client.create_agent(
        instructions=(
            "You are a helpful assistant. The user has declined to see alternative hotel destinations. "
            "Create a polite cancellation message. "
            "Return JSON with: status (should be 'cancelled'), and message (a friendly acknowledgment). "
            "Keep the message brief and understanding."
        ),
        response_format=CancellationMessage,
    ),
    id="cancellation_agent",
)

# NEW: RequestInfoExecutor - pauses workflow to gather human input
request_info_executor = RequestInfoExecutor(id="request_info")

# NEW: DecisionManager instance - routes based on human feedback
decision_manager = DecisionManager(id="decision_manager")

display(
    HTML("""
    <div style='padding: 15px; background: #e3f2fd; border-left: 4px solid #2196f3; border-radius: 4px; margin: 10px 0;'>
        <strong>вң… Created Workflow Components:</strong>
        <ul style='margin: 10px 0 0 0;'>
            <li><strong>availability_agent</strong> - Checks availability with hotel_booking tool</li>
            <li><strong>confirmation_agent</strong> рҹҶ• - Prepares human confirmation request</li>
            <li><strong>alternative_agent</strong> - Suggests alternative cities</li>
            <li><strong>booking_agent</strong> - Encourages booking</li>
            <li><strong>cancellation_agent</strong> рҹҶ• - Handles user declining alternatives</li>
            <li><strong>request_info_executor</strong> рҹҶ• - Pauses workflow for human input</li>
            <li><strong>decision_manager</strong> рҹҶ• - Routes based on human response</li>
        </ul>
    </div>
""")
)
````

## бһҮбҹҶбһ бһ¶бһ“бһ‘бһё 9: бһҹбһ¶бһ„бһҹбһ„бҹӢбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ (Workflow) бһҮбһ¶бһҳбһҪбһҷбһҖбһ¶бһҡбһ…бһјбһӣбһҡбһҪбһҳбһҡбһ”бһҹбҹӢбһҳбһ“бһ»бһҹбҹ’бһҹ

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҷбһҫбһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбҹ’бһҡбһ¶бһ бҹ’бһңбһҡбһ”бһҹбҹӢбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ (workflow) бһҮбһ¶бһҳбһҪбһҷбһ“бһ№бһ„ **бһҖбһ¶бһҡбһ”бһүбҹ’бһҮбһјбһ“бһҸбһ¶бһҳбһӣбһҖбҹ’бһҒбһҒбһҺбҹ’бһҢ** бһҠбҹӮбһӣбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһ•бҹ’бһӣбһјбһңбһҠбҹӮбһӣбһҳбһ¶бһ“бһҖбһ¶бһҡбһ…бһјбһӣбһҡбһҪбһҳбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹ:

**бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһ“бҹғ Workflow:**
```
availability_agent (START)
        вҶ“
   Evaluate conditions
        вҶҷ                    вҶҳ
[no_availability]        [has_availability]
        вҶ“                        вҶ“
confirmation_agent          booking_agent
        вҶ“                        вҶ“
prepare_human_request      display_result
        вҶ“
request_info_executor (PAUSE)
        вҶ“
decision_manager
   вҶҷ         вҶҳ
[yes]        [no]
   вҶ“           вҶ“
alternative  cancellation
   вҶ“           вҶ“
display_result
```

**бһ…бҹҶбһҺбһ»бһ…бһҹбҹҶбһҒбһ¶бһ“бҹӢбҹ—:**
- `availability_agent вҶ’ confirmation_agent` (бһ–бҹҒбһӣбһҠбҹӮбһӣбһҳбһ·бһ“бһҳбһ¶бһ“бһ”бһ“бҹ’бһ‘бһ”бҹӢ)
- `confirmation_agent вҶ’ prepare_human_request` (бһ”бһҳбҹ’бһӣбҹӮбһ„бһ”бҹ’бһҡбһ—бҹҒбһ‘)
- `prepare_human_request вҶ’ request_info_executor` (бһҡбһ„бҹӢбһ…бһ¶бҹҶбһҳбһ“бһ»бһҹбҹ’бһҹ)
- `request_info_executor вҶ’ decision_manager` (бһҸбҹӮбһ„бһҸбҹӮ - бһ•бҹ’бһҸбһӣбҹӢ RequestResponse)
- `decision_manager вҶ’ alternative_agent` (бһ–бҹҒбһӣбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһ“бһ·бһҷбһ¶бһҷ "yes")
- `decision_manager вҶ’ cancellation_agent` (бһ–бҹҒбһӣбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһ“бһ·бһҷбһ¶бһҷ "no")
- `availability_agent вҶ’ booking_agent` (бһ–бҹҒбһӣбһҠбҹӮбһӣбһҳбһ¶бһ“бһ”бһ“бҹ’бһ‘бһ”бҹӢ)
- бһӮбҹ’бһҡбһ”бҹӢбһ•бҹ’бһӣбһјбһңбһ”бһүбҹ’бһ…бһ”бҹӢбһ“бҹ… `display_result`


````python
# Build the workflow with human-in-the-loop routing
workflow = (
    WorkflowBuilder()
    .set_start_executor(availability_agent)
    
    # NO AVAILABILITY PATH (with human-in-the-loop)
    .add_edge(availability_agent, confirmation_agent, condition=no_availability_condition)
    .add_edge(confirmation_agent, prepare_human_request)  # Transform to HumanFeedbackRequest
    .add_edge(prepare_human_request, request_info_executor)  # Send to RequestInfoExecutor
    .add_edge(request_info_executor, decision_manager)    # Always goes to decision manager
    
    # Decision manager routes based on user response
    .add_edge(decision_manager, alternative_agent, condition=user_wants_alternatives_condition)
    .add_edge(decision_manager, cancellation_agent, condition=user_declines_alternatives_condition)
    .add_edge(alternative_agent, display_result)
    .add_edge(cancellation_agent, display_result)
    
    # HAS AVAILABILITY PATH (no human input needed)
    .add_edge(availability_agent, booking_agent, condition=has_availability_condition)
    .add_edge(booking_agent, display_result)
    
    .build()
)

display(
    HTML("""
    <div style='padding: 20px; background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); color: white; border-radius: 8px; margin: 10px 0;'>
        <h3 style='margin: 0 0 15px 0;'>вң… Workflow Built Successfully!</h3>
        <p style='margin: 0; line-height: 1.6;'>
            <strong>Human-in-the-Loop Routing:</strong><br>
            вҖў If <strong>NO availability</strong> вҶ’ confirmation_agent вҶ’ prepare_human_request вҶ’ request_info_executor вҶ’ <strong>PAUSE FOR HUMAN</strong> вҶ’ decision_manager<br>
            &nbsp;&nbsp;вҖў If user says <strong>YES</strong> вҶ’ alternative_agent вҶ’ display_result<br>
            &nbsp;&nbsp;вҖў If user says <strong>NO</strong> вҶ’ cancellation_agent вҶ’ display_result<br>
            вҖў If <strong>availability</strong> вҶ’ booking_agent вҶ’ display_result (no human input needed)
        </p>
    </div>
""")
)
````

## бһҮбҹҶбһ бҹҠбһ¶бһ“ 10: бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ бһҖбһҡбһҺбһёбһҹбһ¶бһҖбһӣбҹ’бһ”бһ„ 1 - бһ‘бһёбһҖбҹ’бһҡбһ»бһ„ бһӮбҹ’бһҳбһ¶бһ“бһ—бһ¶бһ–бһҡбһҖбһ”бһ¶бһ“ (Paris бһҮбһ¶бһҳбһҪбһҷбһҖбһ¶бһҡбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹ)

бһҖбһ¶бһҡбһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһ“бҹҒбҹҮбһ”бһ„бҹ’бһ бһ¶бһүбһ–бһё **бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ–бҹҒбһүбһӣбҹҒбһүбһҠбҹӮбһӣбһҳбһ¶бһ“бһҳбһ“бһ»бһҹбҹ’бһҹбһҖбҹ’бһ“бһ»бһ„бһҹбһ„бһңбһ·бһҗбһё**:

1. бһҹбҹ’бһ“бһҫбһҹбһ»бҹҶбһҹбһҺбҹ’бһӢбһ¶бһӮбһ¶бһҡбһ“бҹ… Paris
2. availability_agent checks вҶ’ бһӮбҹ’бһҳбһ¶бһ“бһ”бһ“бҹ’бһ‘бһ”бҹӢ
3. confirmation_agent creates human-facing question
4. request_info_executor **бһ•бҹ’бһўбһ¶бһҖбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ** бһ бһҫбһҷбһ”бһүбҹ’бһ…бҹҒбһү `RequestInfoEvent`
5. **бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ…бһ¶бһ”бҹӢбһҹбһүбҹ’бһүбһ¶бһ–бҹ’бһҡбһ№бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҺбҹҚ бһ“бһ·бһ„бһҮбһјбһ“бһҠбҹҶбһҺбһ№бһ„бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһ»бһ„бһҹбһјбһӣ**
6. бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһңбһ¶бһҷ "yes" бһ¬ "no"
7. бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ•бҹ’бһүбһҫбһ…бһҳбҹ’бһӣбһҫбһҷбһҸбһ¶бһҳбһҡбһҷбҹҲ `send_responses_streaming()`
8. decision_manager бһҖбҹҶбһҺбһҸбҹӢбһ•бҹ’бһӣбһјбһңбһўбһ“бһ»бһңбһҸбҹ’бһҸбһ‘бҹ…бһҸбһ¶бһҳбһ…бһҳбҹ’бһӣбһҫбһҷ
9. бһӣбһ‘бҹ’бһ’бһ•бһӣбһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷбһ”бһ¶бһ“бһ”бһ„бҹ’бһ бһ¶бһү

**бһӣбҹҶбһ“бһ¶бҹҶбһҹбҹҶбһҒбһ¶бһ“бҹӢбҹҲ**
- бһ”бҹ’бһҡбһҫ `workflow.run_stream()` бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһңбһҠбҹ’бһҸбһҠбҹҶбһ”бһјбһ„
- бһ”бҹ’бһҡбһҫ `workflow.send_responses_streaming(pending_responses)` бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһңбһҠбҹ’бһҸбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢ
- бһҹбҹ’бһҸбһ¶бһ”бҹӢ `RequestInfoEvent` бһҠбһҫбһҳбҹ’бһ”бһёбһҖбҹҶбһҺбһҸбҹӢбһ–бҹҒбһӣбһҠбҹӮбһӣбһ…бһ¶бҹҶбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹ
- бһҹбҹ’бһҸбһ¶бһ”бҹӢ `WorkflowOutputEvent` бһҠбһҫбһҳбҹ’бһ”бһёбһ…бһ¶бһ”бҹӢбһҷбһҖбһӣбһ‘бҹ’бһ’бһ•бһӣбһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷ


````python
display(
    HTML("""
    <div style='padding: 20px; background: #fff3e0; border-left: 4px solid #ff9800; border-radius: 8px; margin: 20px 0;'>
        <h3 style='margin: 0 0 10px 0; color: #e65100;'>рҹ§Ә TEST CASE 1: Paris (No Availability - Human-in-the-Loop)</h3>
        <p style='margin: 0;'>Expected workflow path: availability_agent вҶ’ confirmation_agent вҶ’ request_info_executor вҶ’ <strong>PAUSE</strong> вҶ’ decision_manager вҶ’ (depends on user input)</p>
    </div>
""")
)

# Create request for Paris
request_paris = AgentExecutorRequest(
    messages=[ChatMessage(Role.USER, text="I want to book a hotel in Paris")], 
    should_respond=True
)

# Human-in-the-loop execution pattern
pending_responses: dict[str, str] | None = None
completed = False
workflow_output: str | None = None

print("\nрҹ”„ Starting human-in-the-loop workflow...")
print("=" * 60)

while not completed:
    # First iteration uses run_stream with the request
    # Subsequent iterations use send_responses_streaming with collected human responses
    if pending_responses:
        print(f"\nрҹ“Ө Sending human responses: {pending_responses}")
        stream = workflow.send_responses_streaming(pending_responses)
        pending_responses = None  # Clear immediately after sending
    else:
        print(f"\nрҹҡҖ Starting workflow with request: 'I want to book a hotel in Paris'")
        stream = workflow.run_stream(request_paris)
    
    # Collect all events from this iteration
    events = [event async for event in stream]
    
    # Process events
    requests: list[tuple[str, str]] = []  # (request_id, prompt)
    
    for event in events:
        # Check for human input requests
        if isinstance(event, RequestInfoEvent) and isinstance(event.data, HumanFeedbackRequest):
            print(f"\nвҸёпёҸ  WORKFLOW PAUSED - Human input requested!")
            print(f"   Request ID: {event.request_id}")
            print(f"   Destination: {event.data.destination}")
            requests.append((event.request_id, event.data.prompt))
        
        # Check for workflow outputs
        elif isinstance(event, WorkflowOutputEvent):
            workflow_output = str(event.data)
            completed = True
            print(f"\nвң… Workflow completed with output!")
    
    # If we have human requests, prompt the user
    if requests and not completed:
        responses: dict[str, str] = {}
        for req_id, prompt in requests:
            print(f"\n{'='*60}")
            print(f"рҹ’¬ QUESTION FOR YOU:")
            print(f"   {prompt}")
            print(f"{'='*60}")
            
            # Get user input (in notebook, this will pause execution)
            answer = input("рҹ‘ү Enter 'yes' or 'no': ").strip().lower()
            
            print(f"\nрҹ“қ You answered: {answer}")
            responses[req_id] = answer
        
        pending_responses = responses

print(f"\n{'='*60}")
print(f"рҹҸҶ FINAL WORKFLOW OUTPUT:")
print(f"{'='*60}")

# Display final result
if workflow_output:
    # Try to parse as JSON for pretty display
    try:
        result_data = json.loads(workflow_output)
        if "alternative_destination" in result_data:
            result_obj = AlternativeResult.model_validate_json(workflow_output)
            display(
                HTML(f"""
                <div style='padding: 25px; background: linear-gradient(135deg, #FFD700 0%, #FFA500 100%); border-radius: 12px; box-shadow: 0 4px 12px rgba(255,165,0,0.3); margin: 20px 0;'>
                    <h3 style='margin: 0 0 15px 0; color: #333;'>рҹҸҶ WORKFLOW RESULT</h3>
                    <div style='background: white; padding: 20px; border-radius: 8px;'>
                        <p style='margin: 0 0 10px 0; font-size: 16px;'><strong>Status:</strong> вқҢ No rooms in Paris</p>
                        <p style='margin: 0 0 10px 0; font-size: 16px;'><strong>User Decision:</strong> вң… Accepted alternatives</p>
                        <p style='margin: 0 0 10px 0; font-size: 16px;'><strong>Alternative Suggestion:</strong> рҹҸЁ {result_obj.alternative_destination}</p>
                        <p style='margin: 0; font-size: 14px; color: #666;'><strong>Reason:</strong> {result_obj.reason}</p>
                    </div>
                </div>
            """)
            )
        else:
            # User declined
            display(
                HTML(f"""
                <div style='padding: 25px; background: linear-gradient(135deg, #f44336 0%, #e91e63 100%); color: white; border-radius: 12px; box-shadow: 0 4px 12px rgba(244,67,54,0.3); margin: 20px 0;'>
                    <h3 style='margin: 0 0 15px 0;'>рҹҸҶ WORKFLOW RESULT</h3>
                    <div style='background: white; color: #333; padding: 20px; border-radius: 8px;'>
                        <p style='margin: 0 0 10px 0; font-size: 16px;'><strong>Status:</strong> вқҢ No rooms in Paris</p>
                        <p style='margin: 0 0 10px 0; font-size: 16px;'><strong>User Decision:</strong> рҹҡ« Declined alternatives</p>
                        <p style='margin: 0; font-size: 14px; color: #666;'><strong>Result:</strong> Booking request cancelled</p>
                    </div>
                </div>
            """)
            )
    except:
        print(workflow_output)
````

## бһҮбҹҶбһ бһ¶бһ“ 11: бһҡбһҸбҹӢбһҖбһҡбһҺбһёбһҹбһ¶бһҖбһӣбҹ’бһ”бһ„ 2 - бһ‘бһёбһҖбҹ’бһҡбһ»бһ„ бһҠбҹӮбһӣбһҳбһ¶бһ“бһ”бһ“бҹ’бһ‘бһ”бҹӢ (Stockholm - бһҳбһ·бһ“бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹ)

бһҸбҹҒбһҹбҹ’бһҸбһ“бҹҒбҹҮбһ”бһ„бҹ’бһ бһ¶бһүбһ–бһё **бһ•бҹ’бһӣбһјбһңвҖӢбһҠбҹ„бһҷбһ•бҹ’бһ‘бһ¶бһӣбҹӢ** бһ“бҹ…бһ–бҹҒбһӣбһҳбһ¶бһ“бһ”бһ“бҹ’бһ‘бһ”бҹӢ:

бһӣбҹҶбһ бһјбһҡбһҖбһ¶бһҡбһ„бһ¶бһҡбһ“бҹҒбҹҮбһӣбһҫбһҖбһӣбҹӮбһ„бһ•бҹ’бһӣбһјбһңбһҠбҹӮбһӣбһҳбһ¶бһ“бһҖбһ¶бһҡбһ…бһјбһӣбһҡбһҪбһҳбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹбһ‘бһ¶бҹҶбһ„бһҹбҹ’бһҡбһ»бһ„ бһ“бҹ…бһ–бҹҒбһӣбһҳбһ¶бһ“бһ”бһ“бҹ’бһ‘бһ”бҹӢбҹ”

## бһ…бҹҶбһҺбһ»бһ…бһҹбҹҶбһҒбһ¶бһ“бҹӢ бһ“бһ·бһ„бһўбһ“бһ»бһңбһҸбҹ’бһҸбһӣбҹ’бһўбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһҹбһҳбҹ’бһҡбһ”бһҹбһҳбҹ’бһҡбһҪбһӣбһҮбһ¶бһҳбһҪбһҷбһҳбһ“бһ»бһҹбҹ’бһҹ (Human-in-the-Loop)

### вң… бһўбҹ’бһңбһёбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһҡбҹҖбһ“бҹ–

#### 1. **RequestInfoExecutor бһӮбҹҶбһҡбһј**
бһӣбҹҶбһ“бһ¶бҹҶ human-in-the-loop бһҖбҹ’бһ“бһ»бһ„ Microsoft Agent Framework бһ”бҹ’бһҡбһҫбһ’бһ¶бһҸбһ»бһҹбҹҶбһҒбһ¶бһ“бҹӢбһ”бһёбһ“бҹҒбҹҮбҹ–
- `RequestInfoExecutor` - бһҡбһ¶бһҡбһ¶бҹҶбһ„бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ бһ“бһ·бһ„бһ”бһүбҹ’бһ…бҹҒбһүбһ–бҹ’бһҡбһ№бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҺбҹҚ
- `RequestInfoMessage` - бһҗбҹ’бһ“бһ¶бһҖбҹӢбһҳаё№аёҘбһҠбҹ’бһӢбһ¶бһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ payload бһҳбҹүбһ¶бһҹбҹҠбһёбһ“бһҹбһҳбһёбһҖбһ¶бһҡ (бһҹбһјбһҳ subclass бһ“бҹҒбҹҮ!)
- `RequestResponse` - бһҹбҹ’бһңбҹӮбһ„бһҡбһҪбһҳбһ…бҹҶбһҺбһ„бһ‘бҹҶбһ“бһ»бһҖбһ…бһ·бһҸбҹ’бһҸбһ…бһҳбҹ’бһӣбһҫбһҷбһҳбһ“бһ»бһҹбҹ’бһҹбһҮбһ¶бһҳбһҪбһҷбһҹбҹҶбһҺбһҫбһҠбһҫбһҳ

**бһҖбһ¶бһҡбһҷбһӣбҹӢбһҠбһ№бһ„бһҹбҹҶбһҒбһ¶бһ“бҹӢбҹ–**
- `RequestInfoExecutor` бһҳбһ·бһ“бһ”бҹ’бһҡбһҳбһјбһӣаёӮа№үаёӯаёЎаё№аёҘбһҸбҹ’бһҡбһ№бһҳбһҸбҹӮбһҳбҹ’бһҠбһ„бһ‘бҹҒ - бһңбһ¶бһ–бҹ’бһҡбһ„бһҫбһҷбһҸбҹӮбһ•бҹ’бһўбһ¶бһҖбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ
- бһҖбһјбһҠбһҖбһҳбҹ’бһҳбһ“бҹғбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҸбҹӮбһҹбҹ’бһҠбһ¶бһ”бҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ `RequestInfoEvent` бһ бһҫбһҷбһ”бҹ’бһҡбһҳбһјбһӣбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣ
- бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһ бҹ… `send_responses_streaming()` бһҮбһ¶бһҳбһҪбһҷ dict бһҠбҹӮбһӣбһ•бҹ’бһӮбһјбһ•бҹ’бһӮбһ„ `request_id` бһ‘бҹ…бһ“бһ№бһ„бһ…бһҳбҹ’бһӣбһҫбһҷбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ

#### 2. **бһӣбҹҶбһ“бһ¶бҹҶбһўбһ“бһ»бһңбһҸбҹ’бһҸбһҖбҹ’бһ“бһ»бһ„бһҡбһ”бҹҖбһ”бһҹбҹ’бһ‘бҹ’бһҡбһёбһҳ**
```python
# бһңбһҠбҹ’бһҸбһ‘бһёбһҳбһҪбһҷ
stream = workflow.run_stream(initial_request)

# бһңбһҠбҹ’бһҸбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢ (бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһҠбҹ„бһҷбһҳбһ“бһ»бһҹбҹ’бһҹ)
stream = workflow.send_responses_streaming(pending_responses)

# бһҸбҹӮбһ„бһҸбҹӮбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ–бҹ’бһҡбһ№бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҺбҹҚ
events = [event async for event in stream]
```

#### 3. **бһҹбҹ’бһҗбһ¶бһ”бһҸбҹ’бһҷбһҖбһҳбҹ’бһҳбһҠбҹ„бһҷбһ–бҹ’бһҡбһ№бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҺбҹҚ**
бһҹбҹ’бһҠбһ¶бһ”бҹӢбһ–бҹ’бһҡбһ№бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҺбҹҚбһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢбһҠбһҫбһҳбҹ’бһ”бһёбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбҹ–
- `RequestInfoEvent` - бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹ (бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ•бҹ’бһўбһ¶бһҖ)
- `WorkflowOutputEvent` - бһӣбһ‘бҹ’бһ’бһ•бһӣбһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷбһҳбһ¶бһ“бһ“бҹ… (бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҹбһҳбҹ’бһҡбҹҒбһ…)
- `WorkflowStatusEvent` - бһҖбһ¶бһҡбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡбһ‘бһёбһҸбһ¶бҹҶбһ„ (IN_PROGRESS, IDLE_WITH_PENDING_REQUESTS, бһҮбһ¶бһҠбһҫбһҳ)

#### 4. **Executor бһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“ бһҮбһ¶бһҳбһҪбһҷ @handler**
`DecisionManager` бһ”бһ„бҹ’бһ бһ¶бһүбһ–бһёбһҡбһ”бҹҖбһ”бһ”бһ„бҹ’бһҖбһҫбһҸ executors бһҠбҹӮбһӣбҹ–
- бһ”бҹ’бһҡбһҫ decorator `@handler` бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһҫбһҖбһ”бһ„бҹ’бһ бһ¶бһүбһҳбҹҒбһҸбһјбһҠбһҮбһ¶вҖӢбһҮбҹҶбһ бһ¶бһ“бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ
- бһ‘бһ‘бһҪбһӣбһҹбһ¶бһҡ typed messages (бһ§. `RequestResponse[HumanFeedbackRequest, str]`)
- бһ“бһ¶бҹҶбһ•бҹ’бһӣбһјбһңбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҠбҹ„бһҷбһ•бҹ’бһүбһҫбһҹбһ¶бһҡвҖӢбһ‘бҹ… Executors бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸ
- бһ…бһјбһӣбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ‘бһҹбҹ’бһҹбһ“бһ¶бһ”бһҡбһ·бһ”бһ‘бһҸбһ¶бһҳ `WorkflowContext`

#### 5. **бһҖбһ¶бһҡбһ”бһүбҹ’бһҮбһјбһ“бһҠбҹ„бһҷбһӣбһҖбҹ’бһҒбһҒбһҺбҹ’бһҢ бһҮбһ¶бһҳбһҪбһҷбһҹбҹҒбһ…бһҖбҹ’бһҸбһёбһҹбһҳбҹ’бһҡбҹҒбһ…бһҡбһ”бһҹбҹӢбһҳбһ“бһ»бһҹбҹ’бһҹ**
бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸ function бһӣбһҖбҹ’бһҒбһҒбһҺбҹ’бһҢбһҠбҹӮбһӣбһңбһ¶бһҷбһҸбһҳбҹ’бһӣбҹғбһ…бһҳбҹ’бһӣбһҫбһҷбһҳбһ“бһ»бһҹбҹ’бһҹбҹ–
```python
def user_wants_alternatives_condition(message: Any) -> bool:
    response_text = message.agent_run_response.text.lower()
    return response_text == "yes"
```

### рҹҺҜ бһҖбһ¶бһҡбһўбһ“бһ»бһңбһҸбҹ’бһҸбһҖбҹ’бһ“бһ»бһ„бһ–бһ·бһ—бһ–бһ–бһ·бһҸ:

1. **бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡвҖӢбһўбһ“бһ»бһҳбҹҗбһҸ**
   - бһ‘бһ‘бһҪбһӣбһҖбһ¶бһҡбһўбһ“бһ»бһҳбҹҗбһҸбһ–бһёбһўбҹ’бһ“бһҖбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„ бһҳбһ»бһ“бһ–бҹҒбһӣбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҡбһ”бһ¶бһҷбһҖбһ¶бһҡбһҺбҹҚбһ…бҹҶбһҺбһ¶бһҷ
   - бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҠбҹ„бһҷбһҳбһ“бһ»бһҹбҹ’бһҹ бһҳбһ»бһ“бһҖбһ¶бһҡбһ•бҹ’бһүбһҫбһўбҹҠбһёбһҳбҹӮбһӣбһҹбҹ’бһңбҹҗбһҷбһ”бҹ’бһҡбһңбһҸбҹ’бһҸбһ·
   - бһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһҖбһҡбһҺбһёбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҠбһҠбҹӮбһӣбһҳбһ¶бһ“бһҸбһҳбҹ’бһӣбҹғбһҒбҹ’бһ–бһҹбҹӢ бһҳбһ»бһ“бһўбһ“бһ»бһңбһҸбҹ’бһҸ

2. **бһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбһ¶бһҸбһ·бһҖбһ¶**
   - бһ•бҹ’бһ‘бһ»бҹҮбһҳбһ¶бһҸбһ·бһҖбһ¶бһҠбҹӮбһӣбһ…бһ¶бһ”бҹӢбһўбһ¶бһҡбһҳбҹ’бһҳбһҺбҹҚбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҠбҹ„бһҷбһҳбһ“бһ»бһҹбҹ’бһҹ
   - бһҹбҹ’бһ“бһҫбһұбҹ’бһҷ Moderator бһ’бҹ’бһңбһҫбһҹбҹҒбһ…бһҖбҹ’бһҸбһёбһҹбһҳбҹ’бһҡбҹҒбһ…бһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷбһӣбһҫбһҖбһҡбһҺбһёбһҠбҹӮбһ“бһҖбҹҶбһҺбһҸбҹӢ
   - бһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҡбҹҶбһӣбҹ’бһңбһҖбһ‘бҹ…бһҳбһ“бһ»бһҹбҹ’бһҹбһ–бҹҒбһӣбһҠбҹӮбһӣбһҖбһ¶бһҡбһ‘бҹҶбһ“бһ»бһҖбһ…бһ·бһҸбҹ’бһҸбһҡбһ”бһҹбҹӢ AI бһ‘бһ¶бһ”

3. **бһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳбһўбһҸбһ·бһҗбһ·бһҮбһ“**
   - бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһұбҹ’бһҷ AI бһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһҹбҹҶбһҺбһҪбһҡбһ’бһҳбҹ’бһҳбһҸбһ¶бһҠбҹ„бһҷбһҹбҹ’бһңбҹҗбһҷбһ”бҹ’бһҡбһңбһҸбҹ’бһҸбһ·
   - бһҡбҹҶбһӣбһ„бһ”бһүбҹ’бһ бһ¶бһҹбҹ’бһҳбһ»бһӮбһҹбҹ’бһҳбһ¶бһүбһ‘бҹ…бһҸбҹҶбһҺбһ¶бһ„бһҳбһ“бһ»бһҹбҹ’бһҹ
   - бһҹбһҪбһҡбһўбһҸбһ·бһҗбһ·бһҮбһ“бһҗбһ¶бһ–бһҪбһҖбһӮбҹҒбһ…бһ„бҹӢбһ“бһ·бһҷбһ¶бһҷбһҮбһ¶бһҳбһҪбһҷбһҳбһ“бһ»бһҹбҹ’бһҹбһ¬бһ‘бҹҒ

4. **бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡвҖӢбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ**
   - бһҹбһҪбһҡбһҳбһ“бһ»бһҹбҹ’бһҹбһҠбһҫбһҳбҹ’бһ”бһёбһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһ”бһүбҹ’бһ бһ¶бһ…бһҳбҹ’бһҡбһ¶бһҹбҹӢбһ“бҹғбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ
   - бһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһҖбһ¶бһҡбһ”бһҖбһҹбҹ’бһҡбһ¶бһҷбһҡбһ”бһҹбҹӢ AI бһӣбһҫбһҜбһҖбһҹбһ¶бһҡбһҳбһ·бһ“бһ…бҹ’бһ”бһ¶бһҹбҹӢ
   - бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһұбҹ’бһҷбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһҮбҹ’бһҡбһҫбһҹбһҡбһңбһ¶бһ„бһҖбһ¶бһҡбһ”бһҖбһҹбҹ’бһҡбһ¶бһҷбһ”бҹүбҹҮбһ–бһ¶бһӣбҹӢбһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһҸбҹ’бһҡбһ№бһҳбһҸбҹ’бһҡбһјбһң

5. **бһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһҹбҹҶбһҒбһ¶бһ“бҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҹбһ»бһңбһҸбҹ’бһҗбһ·бһ—бһ¶бһ–**
   - бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹбһҳбһ»бһ“бһҹбһҖбһҳбҹ’бһҳбһ—бһ¶бһ–бһҠбҹӮбһӣбһҳбһ·бһ“бһўбһ¶бһ…бһӣбһ»бһ”бһ”бһ¶бһ“
   - бһ‘бһ‘бһҪбһӣбһҖбһ¶бһҡбһўбһ“бһ»бһҳбҹҗбһҸбһҳбһ»бһ“бһ…бһјбһӣбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡвҖӢбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҹбҹҶбһ„бһ¶бһҸбҹӢ
   - бһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһҹбҹҒбһ…бһҖбҹ’бһҸбһёбһҹбһҳбҹ’бһҡбҹҒбһ…бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ§бһҹбҹ’бһҹбһ¶бһ бһҖбһҳбҹ’бһҳбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһӮбҹ„бһҡбһ–бһ…бҹ’бһ”бһ¶бһ”бҹӢ (бһҹбһ»бһҒбһ¶бһ—бһ·бһ”бһ¶бһӣ, бһ бһ·бһҡбһүбҹ’бһүбһңбһҸбҹ’бһҗбһ»)

6. **бһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡвҖӢбһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳ**
   - бһҹбһ¶бһ„бһҹбһ„бҹӢбһ”бһ»бһҸбҹӢбһҖбһҮбһҮбҹӮбһҖбһҠбҹӮбһӣбһҹбһҪбһҡбһ–бһ„бҹ’бһҡбһёбһҖбһҹбҹҶбһҺбһҪбһҡбһ”бһ“бҹ’бһҗбҹӮбһҳ
   - бһ”бһ„бҹ’бһҖбһҫбһҸбһңбһёбһҹбһ¶бһҠбҹӮбһӣбһҺбҹӮбһ“бһ¶бҹҶбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһҸбһ¶бһҳбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҹбҹ’бһҳбһ»бһӮбһҹбҹ’бһҳбһ¶бһү
   - бһҡбһ…бһ“бһ¶бһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһҠбҹӮбһӣбһҹбһ бһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷбһҳбһ“бһ»бһҹбҹ’бһҹбһҮбһ¶бһ‘бһҳбҹ’бһҡбһ„бҹӢбһҮбҹҶбһ бһ¶бһ“бҹӢбҹ—

### рҹ”„ бһҖбһ¶бһҡбһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бҹ– бһҳбһ¶бһ“бһҳбһ“бһ»бһҹбҹ’бһҹбһҖбҹ’бһ“бһ»бһ„бһӣбҹҶбһ“бһ¶бҹҶ бһ‘бһӣбҹӢбһ“бһ№бһ„бһӮбҹ’бһҳбһ¶бһ“

| Feature | Conditional Workflow | Human-in-the-Loop Workflow |
|---------|---------------------|---------------------------|
| **Execution** | Single `workflow.run()` | Loop with `run_stream()` + `send_responses_streaming()` |
| **User Input** | None (fully automated) | Interactive prompts via `input()` or UI |
| **Components** | Agents + Executors | + RequestInfoExecutor + DecisionManager |
| **Events** | AgentExecutorResponse only | RequestInfoEvent, WorkflowOutputEvent, etc. |
| **Pausing** | No pausing | Workflow pauses at RequestInfoExecutor |
| **Human Control** | No human control | Humans make key decisions |
| **Use Case** | Automation | Collaboration & oversight |

### рҹҡҖ бһӣбҹҶбһ“бһ¶бҹҶбһҖбһҳбҹ’бһҡбһ·бһҸбһҒбҹ’бһ–бһҹбҹӢ:

#### бһ…бҹҶбһҺбһ»бһ…бһҹбҹҒбһ…бһҖбҹ’бһҸбһёбһҹбһҳбҹ’бһҡбҹҒбһ…бһҳбһ“бһ»бһҹбҹ’бһҹбһ…бҹ’бһҡбһҫбһ“
бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҳбһ¶бһ“бһ…бҹҶбһҺбһ»бһ… `RequestInfoExecutor` бһ…бҹ’бһҡбһҫбһ“бһҖбҹ’бһ“бһ»бһ„бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҠбһјбһ…бһӮбҹ’бһ“бһ¶:
```python
.add_edge(agent1, request_info_1)  # бһҖбһ¶бһҡбһҹбһҳбҹ’бһҡбҹҒбһ…бһ…бһ·бһҸбҹ’бһҸбһҠбҹҶбһ”бһјбһ„бһҡбһ”бһҹбҹӢбһҳбһ“бһ»бһҹбҹ’бһҹ
.add_edge(decision_manager_1, agent2)
.add_edge(agent2, request_info_2)  # бһҖбһ¶бһҡбһҹбһҳбҹ’бһҡбҹҒбһ…бһ…бһ·бһҸбҹ’бһҸбһ‘бһёбһ–бһёбһҡбһҡбһ”бһҹбҹӢбһҳбһ“бһ»бһҹбҹ’бһҹ
.add_edge(decision_manager_2, final_agent)
```

#### бһҖбһ¶бһҡбһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһ–бҹҒбһӣбһ•бһ»бһҸбһҖбҹҶбһҺбһҸбҹӢ
бһўбһ“бһ»бһңбһҸбҹ’бһҸбһҖбһ¶бһҡбһҖбҹҶбһҺбһҸбҹӢбһ–бҹҒбһӣбһ•бһ»бһҸбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ…бһҳбҹ’бһӣбһҫбһҷбһҳбһ“бһ»бһҹбҹ’бһҹ:
```python
import asyncio

try:
    answer = await asyncio.wait_for(
        asyncio.to_thread(input, "Enter yes/no: "),
        timeout=60.0
    )
except asyncio.TimeoutError:
    answer = "no"  # бһӣбҹҶбһ“бһ¶бҹҶбһҠбһҫбһҳбһ‘бҹ…бһҮбһ¶бһҮбһҳбҹ’бһҡбһҫбһҹбһҠбҹӮбһӣбһҳбһ¶бһ“бһҹбһ»бһңбһҸбҹ’бһҗбһ·бһ—бһ¶бһ–
```

#### бһҖбһ¶бһҡбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣ UI бһҹбһҳбҹ’бһ”бһјбһҡ
бһҮбҹҶбһ“бһҪбһҹ `input()`, бһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһҮбһ¶бһҳбһҪбһҷ Web UI, Slack, Teams, бһӣбҹ–
```python
if isinstance(event, RequestInfoEvent):
    # бһ•бҹ’бһүбһҫбһҖбһ¶бһҡбһҮбһјбһ“бһҠбҹҶбһҺбһ№бһ„бһ‘бҹ…бһҶбһ¶бһ“бҹӮбһӣбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһ…бһјбһӣбһ…бһ·бһҸбҹ’бһҸ
    await slack_client.send_message(
        user_id=current_user,
        text=event.data.prompt,
        request_id=event.request_id
    )
```

#### Conditional Human-in-the-Loop
бһ“бһ·бһҷбһ¶бһҷбһ”бҹ’бһҡбһ¶бһ”бҹӢбһҸбҹӮбһҹбҹ’бһ“бһҫбһҹбһҪбһҡбһҳбһ“бһ»бһҹбҹ’бһҹбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢбһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮ:
```python
def needs_human_approval_condition(message: Any) -> bool:
    # бһ•бҹ’бһүбһҫбһ‘бҹ…бһҳбһ“бһ»бһҹбҹ’бһҹбһҸбҹӮбһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮ бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһҖбһҳбҹ’бһҡбһ·бһҸбһҮбһҝбһҮбһ¶бһҖбҹӢбһ‘бһ¶бһ” бһ¬ бһҸбһҳбҹ’бһӣбҹғбһҒбҹ’бһ–бһҹбҹӢ
    if result.confidence < 0.7 or result.value > 10000:
        return True
    return False
```

### вҡ пёҸ бһҖбһ¶бһҡбһўбһ“бһ»бһңбһҸбҹ’бһҸбһӣбҹ’бһўбһ”бҹҶбһ•бһ»бһҸбҹ–

1. **бһҸбҹӮбһ„бһҸбҹӮбһ”бһ„бҹ’бһҖбһҫбһҸбһҗбҹ’бһ“бһ¶бһҖбҹӢбһ…бһҳбҹ’бһӣбһ„бһ–бһё `RequestInfoMessage`**
   - бһ•бҹ’бһҠбһӣбҹӢбһҹбһ»бһңбһҸбҹ’бһҗбһ·бһ—бһ¶бһ–бһ”бҹ’бһҡбһ—бҹҒбһ‘ бһ“бһ·бһ„бһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷ
   - бһўбһ¶бһ…бһ•бҹ’бһҠбһӣбҹӢбһ”бһҡбһ·бһ”бһ‘бһҹбһҳбҹ’бһ”бһјбһҡбһ”бҹӮбһ”бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһҡбҹҖбһ”бһ…бҹҶ UI
   - бһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһ…бҹҒбһҸбһ“бһ¶вҖӢбһ“бҹғбһ”бҹ’бһҡбһ—бҹҒбһ‘бһҹбҹҶбһҺбһҫбһ“бһёбһҳбһҪбһҷбҹ—

2. **бһ”бҹ’бһҡбһҫбһҹбҹҶбһҺбһҪбһҡбһ–бһҺбҹҢбһ“бһ¶бһҠбҹӮбһӣбһ…бҹ’бһ”бһ¶бһҹбҹӢ**
   - бһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһ”бһҡбһ·бһ”бһ‘бһўбҹҶбһ–бһёбһўбҹ’бһңбһёбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһҖбҹҶбһ–бһ»бһ„бһҹбһҪбһҡ
   - бһ–бһ“бҹ’бһҷбһӣбҹӢбһ•бһӣбһ”бҹүбҹҮбһ–бһ¶бһӣбҹӢбһ“бҹғбһҮбһҳбҹ’бһҡбһҫбһҹбһ“бһёбһҳбһҪбһҷбҹ—
   - бһ…бһ„бҹӢбһұбҹ’бһҷбһҹбҹҶбһҺбһҪбһҡбһҮбһ¶бһҖбҹӢбһҹбҹ’бһҸбҹӮбһ„ бһ“бһ·бһ„бһ„бһ¶бһҷбһҷбһӣбҹӢ

3. **бһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһҳбһ·бһ“бһҡбҹҶбһ–бһ№бһ„бһ‘бһ»бһҖ**
   - бһ•бҹ’бһ‘бҹҖбһ„бһ•бҹ’бһ‘бһ¶бһҸбҹӢбһ…бһҳбҹ’бһӣбһҫбһҷбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ
   - бһ•бҹ’бһҸбһӣбҹӢбһҸбһҳбҹ’бһӣбҹғбһӣбҹҶбһ“бһ¶бҹҶбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһҳбһ·бһ“бһҸбҹ’бһҡбһ№бһҳбһҸбҹ’бһҡбһјбһң
   - бһ•бҹ’бһҠбһӣбҹӢбһҹбһ¶бһҡ бһ”бһ„бҹ’бһ бһ¶бһүбһҖбҹҶбһ бһ»бһҹбһҷбҹүбһ¶бһ„бһ…бҹ’бһ”бһ¶бһҹбҹӢ

4. **бһҸбһ¶бһҳбһҠбһ¶бһ“ Request IDs**
   - бһ”бҹ’бһҡбһҫбһҖбһ¶бһҡбһ…бһ„бһҖбҹ’бһҡбһ„бһҡбһңбһ¶бһ„ `request_id` бһ“бһ·бһ„бһ…бһҳбҹ’бһӣбһҫбһҷ
   - бһҖбһ»бҹҶбһ–бҹ’бһҷбһ¶бһҷбһ¶бһҳбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бһҠбҹ„бһҷбһҠбҹғ

5. **бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҳбһ·бһ“бһ”бһ·бһ‘бһ”бһ¶бҹҶбһ„**
   - бһҖбһ»бҹҶбһ”бһ·бһ‘бһҗбҹ’бһҳбҹ„бһ„бһҠбҹӮбһӣбһҡбһ„бҹӢбһ…бһ¶бҹҶбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣ
   - бһ”бҹ’бһҡбһҫбһӣбҹҶбһ“бһ¶бҹҶ async бһ‘бһ¶бҹҶбһ„бһҳбһјбһӣ
   - бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһҖбһ¶бһҡбһҡбһҸбҹӢбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹ„бһҷбһҹбһҖбһӣ

### рҹ“ҡ бһӮбҹҶбһ“бһ·бһҸбһ–бһ¶бһҖбҹӢбһ–бҹҗбһ“бҹ’бһ’:

- **Agent Middleware** - бһ’бҹ’бһңбһҫбһўбһ“бҹ’бһҸбһҡбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ agent calls (бһҹбҹҖбһңбһ—бҹ…бһҖбҹҶбһҺбһҸбҹӢбһҸбҹ’бһҡбһ¶бһҳбһ»бһ“)
- **Workflow State Management** - бһҡбһҖбҹ’бһҹбһ¶бһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҡбһңбһ¶бһ„бһҖбһ¶бһҡбһҡбһҸбҹӢ
- **Multi-Agent Collaboration** - бһҹбһҳбҹ’бһҡбһ”бһҹбһҳбҹ’бһҡбһҪбһӣ human-in-the-loop бһҮбһ¶бһҳбһҪбһҷбһҖбҹ’бһҡбһ»бһҳбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡ
- **Event-Driven Architectures** - бһҹбһ¶бһ„бһҹбһ„бҹӢбһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһҠбҹ„бһҷбһ–бҹ’бһҡбһ№бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҺбҹҚ

- вң… бһҡбһ¶бһҡбһ¶бҹҶбһ„бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ бһҠбһҫбһҳбҹ’бһ”бһёбһ”бҹ’бһҡбһҳбһјбһӣбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһ–бһёбһҳбһ“бһ»бһҹбҹ’бһҹ
- вң… бһ”бҹ’бһҡбһҫ `RequestInfoExecutor` бһ“бһ·бһ„ `RequestInfoMessage`
- вң… бһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһҖбһ¶бһҡбһўбһ“бһ»бһңбһҸбҹ’бһҸбһҹбҹ’бһ‘бҹ’бһҡбһёбһҳбһҮбһ¶бһҳбһҪбһҷбһ–бҹ’бһҡбһ№бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҺбҹҚ
- вң… бһ”бһ„бҹ’бһҖбһҫбһҸ executor бһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“ бһҮбһ¶бһҳбһҪбһҷ `@handler`
