# 🔍 Enterprise RAG ជាមួយ Microsoft Foundry (.NET)

## 📋 គោលបំណងការសិក្សា

កំណត់ត្រានេះបង្ហាញពីរបៀបកសាងប្រព័ន្ធ Retrieval-Augmented Generation (RAG) កម្រិតសហគ្រាស ដោយប្រើ Microsoft Agent Framework ក្នុង .NET ជាមួយ Microsoft Foundry។ អ្នកនឹងរៀនបង្កើតភ្នាក់ងាររួចរាល់សម្រាប់ផលិតកម្មដែលអាចស្វែងរកក្នុងឯកសារ និងផ្តល់ចម្លើយដែលមានភាពត្រឹមត្រូវ និងយល់ដឹងពីបរិបទ ជាមួយសុវត្ថិភាពនិងសមត្ថភាពធំទូលាយសម្រាប់សហគ្រាស។

**សមត្ថភាព Enterprise RAG ដែលអ្នកនឹងកសាង៖**
- 📚 **Document Intelligence**: ការបំព្រួចឯកសារដែលមានកម្រិតខ្ពស់ជាមួយសេវាកម្ម Azure AI
- 🔍 **Semantic Search**: ការស្វែងរកវ៉ិចទ័រដែលមានសមត្ថភាពខ្ពស់ជាមួយលក្ខណៈសម្រាប់សហគ្រាស
- 🛡️ **Security Integration**: ការរួមបញ្ចូលសុវត្ថិភាព — ការចូលប្រើផ្អែកលើតួនាទី និងលំនាំការការពារទិន្នន័យ
- 🏢 **Scalable Architecture**: ស្ថาปត្យកម្មដែលអាចជម្រុញ (Scalable) — ប្រព័ន្ធ RAG រួចរាល់សម្រាប់ផលិតកម្ម ជាមួយការត្រួតពិនិត្យ

## 🎯 ស្ថាបត្យកម្ម Enterprise RAG

### ឧបករណ៍ស្នូលសម្រាប់សហគ្រាស
- **Microsoft Foundry**: វេទិកា AI សម្រាប់សហគ្រាសដែលបានគ្រប់គ្រងជាមួយសុវត្ថិភាព និងការអនុក្រឹត
- **Persistent Agents**: ភ្នាក់ងារមានរដ្ឋភាពដែលមានប្រវត្តិការសន្ទនា និងការគ្រប់គ្រងបរិបទ
- **Vector Store Management**: ការគ្រប់គ្រងឃ្លាំងវ៉ិចទ័រ៖ ការកំណត់លេខសញ្ញា និងការទាញយកឯកសារកម្រិតសហគ្រាស
- **Identity Integration**: ការរួមបញ្ចូលអត្តសញ្ញាណ៖ ការផ្ទៀងផ្ទាត់ Azure AD និងការគ្រប់គ្រងការចូលប្រើផ្អែកលើតួនាទី

### អត្ថប្រយោជន៍ .NET សម្រាប់សហគ្រាស
- **Type Safety**: សុវត្ថិភាពប្រភេទ៖ ការផ្ទៀងផ្ទាត់នៅពេលកំប្លុកសម្រាប់ប្រតិបត្តិការនិងរចនាសម្ព័ន្ធទិន្នន័យ RAG
- **Async Performance**: សមត្ថភាព Async៖ ការបំព្រួចឯកសារ និងប្រតិបត្តិការ​ស្វែងរក​ដែលមិនបណ្តាក់
- **Memory Management**: ការគ្រប់គ្រងអង្គចងចាំ៖ ការប្រើប្រាស់ធនធានយ៉ាងមានប្រសិទ្ធភាពសម្រាប់បណ្ណាល័យឯកសារធំៗ
- **Integration Patterns**: លំនាំការរួមបញ្ចូល៖ ការរួមបញ្ចូលសេវាកម្ម Azure ដើមជាមួយ dependency injection

```csharp
Document Upload → Security Validation → Vector Processing → Index Creation
                      ↓                    ↓                  ↓
User Query → Authentication → Semantic Search → Context Ranking → AI Response
```

### ឧបករណ៍ស្នូល .NET
- **Azure.AI.Agents.Persistent**: ការគ្រប់គ្រងភ្នាក់ងារសម្រាប់សហគ្រាសជាមួយការទុកសភាពរដ្ឋ
- **Azure.Identity**: ការផ្ទៀងផ្ទាត់រួមសម្រាប់ការចូលប្រើសេវា Azure ដោយភាពសុវត្ថិ
- **Microsoft.Agents.AI.AzureAI**: ការអនុវត្តគ្រោងភ្នាក់ងារ​ដែលខ្ទង់សម្រាប់ Azure
- **System.Linq.Async**: ប្រតិបត្តិការលីញ (LINQ) អាស៊ីញដែលមានសមត្ថភាពខ្ពស់

## 🔧 លក្ខណៈពិសេស និងអត្ថប្រយោជន៍សម្រាប់សហគ្រាស

### សុវត្ថិភាព និងការអនុក្រឹត
- **Azure AD Integration**: ការរួមបញ្ចូល Azure AD៖ ការគ្រប់គ្រងអត្តសញ្ញាណ និងការផ្ទៀងផ្ទាត់សម្រាប់សហគ្រាស
- **Role-Based Access**: ការចូលប្រើផ្អែកលើតួនាទី៖ សិទ្ធិលម្អិតសម្រាប់ការចូលមើលឯកសារ និងប្រតិបត្តិការ
- **Data Protection**: ការការពារទិន្នន័យ៖ ការបម្រាញ់ទិន្នន័យនៅពេលស្នាក់ និងនៅពេលដឹកជញ្ជូនសម្រាប់ឯកសារមានអារម្មណ៍ភាគច្រើន
- **Audit Logging**: ការកំណត់ហេតុខ្ពស់: ការតាមដានសកម្មភាពយ៉ាងទូលំទូលាយសម្រាប់ការទាមទារ​​អនុក្រឹត

### សមត្ថភាព និងការជម្រុញ
- **Connection Pooling**: ការជម្រះការតភ្ជាប់ (Connection Pooling): ការគ្រប់គ្រងការតភ្ជាប់សេវា Azure យ៉ាងមានប្រសិទ្ធភាព
- **Async Processing**: ប្រតិបត្តិការមិនបណ្តាក់សម្រាប់សេណារីយ៉ូដែលមាន throughput ខ្ពស់
- **Caching Strategies**: យុទ្ធសាស្ត្រកាស្សេ៖ ការកាស្សេយ៉ាងឆ្លាតសម្រាប់ឯកសារដែលចូលប្រើញឹកញាប់
- **Load Balancing**: ការបន្ថែមទំនាញ (Load Balancing): ការប្រតិបត្តិចែកចាយសម្រាប់ការដាក់ប្រេីប្រាស់ក្នុងវិសាលភាពធំ

### ការគ្រប់គ្រង និងការត្រួតពិនិត្យ
- **Health Checks**: ការត្រួតពិនិត្យសុខភាព៖ ការត្រួតពិនិត្យដែលបញ្ចូលរួចសម្រាប់ធាតុប្រព័ន្ធ RAG
- **Performance Metrics**: មាត្រដ្ឋានសមត្ថភាព៖ វិភាគលម្អិតពីគុណភាពការស្វែងរក និងពេលឆ្លើយតប
- **Error Handling**: ការគ្រប់គ្រងកំហុស៖ ការគ្រប់គ្រងករណីលើកលម្អិតជាមួយគោលនយោបាយព្យាយាមឡើងវិញ
- **Configuration Management**: ការគ្រប់គ្រងការកំណត់រចនាសម្ព័ន្ធ៖ ការកំណត់ដោយផ្អែកលើបរិយាកាសជាមួយការផ្ទៀងផ្ទាត់

## 📊 គំរូ Enterprise RAG

### គំរូគ្រប់គ្រងឯកសារ
- **Bulk Upload**: ការផ្ទុកចូលជាក្បាល (Bulk Upload)៖ ការបំព្រួចយ៉ាងមានប្រសិទ្ធភាពសម្រាប់បណ្ណាល័យឯកសារធំ
- **Incremental Updates**: ការអាប់ដេតបន្ដ៖ ការជួយបន្ថែម និងកែប្រែឯកសារនៅពេលពិត
- **Version Control**: ការគ្រប់គ្រងកំណែ៖ ការគ្រប់គ្រងកំណែឯកសារ និងការតាមដានការផ្លាស់ប្តូរ
- **Metadata Management**: ការគ្រប់គ្រងមេតាដាតា៖ លក្ខណៈឯកសារដែលមានព័ត៌មានសម្បូរបែបសម្រាប់ចំណាត់ថ្នាក់

### គំរូស្វែងរក និងទាញយក
- **Hybrid Search**: ស្វែងរកលំឡើង (Hybrid Search)៖ ការរួមបញ្ចូលស្វែងរកសេម៉ង់ទិក និងពាក្យគន្លឹះសម្រាប់លទ្ធផលល្អបំផុត
- **Faceted Search**: ស្វែងរកតាមផ្នែក (Faceted Search)៖ ការបម្រៀប និងការត្រង់ចម្រង់ច្រើនវិមាត្រ
- **Relevance Tuning**: កែតម្រូវភាពសមស្រប៖ អាល់ហ្គរីធម៍កំណត់ពិន្ទុឧបករណ៍សម្រាប់តម្រូវការផ្នែកជាក់លាក់
- **Result Ranking**: ការរៀបលំដាប់លទ្ធផល៖ ការរៀបចំខ្ពស់ជាមួយការរួមបញ្ចូលតុល្យវិស័យអាជីវកម្ម

### គំរូសុវត្ថិភាព
- **Document-Level Security**: សុវត្ថិភាពជាន់ឯកសារ៖ ការគ្រប់គ្រងការចូលប្រើយ៉ាងលម្អិតក្នុងមួយឯកសារ
- **Data Classification**: ចាត់ថ្នាក់ទិន្នន័យ៖ ការស្លាកភាពងាយរងគ្រោះដោយស្វ័យប្រវត្តិ និងការការពារ
- **Audit Trails**: ទិន្នន័យតាមដាន (Audit Trails)៖ ការកំណត់ហេតុខ្ពស់នៃប្រតិបត្តិការទាំងអស់របស់ RAG
- **Privacy Protection**: ការពារភាពឯកជន៖ សមត្ថភាពរកឃើញ PII និងការរាំងសារ

### ការការពារទិន្នន័យ
- **Encryption**: ការអ៊ិនគ្រីបពីចុងទៅចុងសម្រាប់ឯកសារ និងសន្ទស្សន៍ស្វែងរក
- **Access Controls**: ការគ្រប់គ្រងការចូលប្រើ៖ ការរួមបញ្ចូលជាមួយ Azure AD សម្រាប់សិទ្ធិអ្នកប្រើ និងក្រុម
- **Data Residency**: ការគ្រប់គ្រងទីតាំងភូមិសាស្ត្រសម្រាប់ការអនុក្រឹត
- **Backup & Recovery**: ការបម្រុងទុក និងស្តារឡើងវិញ៖ សមត្ថភាពបម្រុងទុកដោយស្វ័យប្រវត្តិ និងស្តារជាករណីគ្រោះមហន្តរការ

### ការគ្រប់គ្រងអង្គចងចាំ

- **Garbage Collection**: ការជ្រើសរើសកំចាត់សំរាម (Garbage Collection)៖ លំនាំចែកចាយអង្គចងចាំដែលបានអុបថម

````polyglot-notebook
using System;
using System.Linq;
using Azure.AI.Agents.Persistent;
using Azure.Identity;
using Microsoft.Agents.AI;
````

````polyglot-notebook
 using DotNetEnv;
````

````polyglot-notebook
Env.Load("../../../.env");
````

````polyglot-notebook
var azure_foundry_endpoint = Environment.GetEnvironmentVariable("AZURE_AI_PROJECT_ENDPOINT") ?? throw new InvalidOperationException("AZURE_AI_PROJECT_ENDPOINT is not set.");
var azure_foundry_model_id = Environment.GetEnvironmentVariable("AZURE_AI_MODEL_DEPLOYMENT_NAME") ?? "gpt-4.1-mini";
````

````polyglot-notebook
string pdfPath = "./document.md";
````

````polyglot-notebook
using System.IO;

async Task<Stream> OpenImageStreamAsync(string path)
{
	return await Task.Run(() => File.OpenRead(path));
}

var pdfStream = await OpenImageStreamAsync(pdfPath);
````

````polyglot-notebook
var persistentAgentsClient = new PersistentAgentsClient(azure_foundry_endpoint, new AzureCliCredential());
````

````polyglot-notebook
PersistentAgentFileInfo fileInfo = await persistentAgentsClient.Files.UploadFileAsync(pdfStream, PersistentAgentFilePurpose.Agents, "demo.md");
````

````polyglot-notebook
PersistentAgentsVectorStore fileStore =
            await persistentAgentsClient.VectorStores.CreateVectorStoreAsync(
                [fileInfo.Id],
                metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
PersistentAgent agentModel = await persistentAgentsClient.Administration.CreateAgentAsync(
            azure_foundry_model_id,
            name: "DotNetRAGAgent",
            tools: [new FileSearchToolDefinition()],
            instructions: """
                You are an AI assistant designed to answer user questions using only the information retrieved from the provided document(s).

                - If a user's question cannot be answered using the retrieved context, **you must clearly respond**: 
                "I'm sorry, but the uploaded document does not contain the necessary information to answer that question."
                - Do not answer from general knowledge or reasoning. Do not make assumptions or generate hypothetical explanations.
                - Do not provide definitions, tutorials, or commentary that is not explicitly grounded in the content of the uploaded file(s).
                - If a user asks a question like "What is a Neural Network?", and this is not discussed in the uploaded document, respond as instructed above.
                - For questions that do have relevant content in the document (e.g., Contoso's travel insurance coverage), respond accurately, and cite the document explicitly.

                You must behave as if you have no external knowledge beyond what is retrieved from the uploaded document.
                """,
            toolResources: new()
            {
                FileSearch = new()
                {
                    VectorStoreIds = { fileStore.Id },
                }
            },
            metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
AIAgent agent = await persistentAgentsClient.GetAIAgentAsync(agentModel.Id);
````

````polyglot-notebook
AgentThread thread = agent.GetNewThread();
````

````polyglot-notebook
Console.WriteLine(await agent.RunAsync("Can you explain Contoso's travel insurance coverage?", thread));
````
